## Пересечение по Тюмени

Импорт данных по Тюмени

```python
import pandas as pd

#данные марта
calls_tmn_march = pd.read_excel('C://Users//ws-tmn-mark-13//Desktop//Пересечение звонков и чатов//Тюмень//Звонки март Авито.xlsx')
chat_tmn_march = pd.read_excel('C://Users//ws-tmn-mark-13//Desktop//Пересечение звонков и чатов//Тюмень//Чаты март Авито.xlsx')

#данные апреля
calls_tmn_april = pd.read_excel('C://Users//ws-tmn-mark-13//Desktop//Пересечение звонков и чатов//Тюмень//Звонки апрель Авито.xlsx')
chat_tmn_april = pd.read_excel('C://Users//ws-tmn-mark-13//Desktop//Пересечение звонков и чатов//Тюмень//Чаты апрель Авито.xlsx')
```

Вывод первых 5 строк таблиц


В таблицах хранятся номера телефонов клиентов, по которым заведены заявки (и с чатов, и со звонков)


Номера приведены к общему виду: без дополнительных знаков и пробелов.


```python
display(calls_tmn_march.head())
display(chat_tmn_march.head())
display(calls_tmn_april.head())
display(chat_tmn_april.head())
```

Информация об анализируемых данных:
- количество столбцов
- количество строк
- тип данных

```python
print('Звонки за март')
print()
calls_tmn_march.info()
print()
print('Чаты за март')
print()
chat_tmn_march.info()
print()
print('Звонки за апрель')
print()
calls_tmn_april.info()
print()
print('Чаты за апрель')
print()
chat_tmn_april.info()
```

**Список номеров клиентов, которые совершили оба действия**

```python
#в марте
cross_tmn_march = list(set(calls_tmn_march['Номер клиента']) & set(chat_tmn_march['Номер клиента']))

#в апреле
cross_tmn_april = list(set(calls_tmn_april['Номер клиента']) & set(chat_tmn_april['Номер клиента']))
```

```python
print(cross_tmn_march)
print()
print(cross_tmn_april)
```

### Количество клиентов, которые совершили оба действия

```python
print('Пересечение в марте')
print(len(cross_tmn_march))
print()
print('Пересечение в апреле')
print(len(cross_tmn_april))
```

```python
share_march_tmn = (151 / (len(calls_tmn_march['Номер клиента']) + len(chat_tmn_march['Номер клиента']) - 2 - 151))
share_april_tmn = (103 / (len(calls_tmn_april['Номер клиента']) + len(chat_tmn_april['Номер клиента']) - 2 - 103))
```

```python
print('Доля пересечения от общего количества заявок в марте')
print("{:.2%}".format(share_march_tmn))
print()
print('Доля пересечения от общего количества заявок в апреле')
print("{:.2%}".format(share_april_tmn))
```

## Сравнение по городам

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
march = [6.72, 8.15, 4.11, 4.86, 3.7, 3.18]
april =[7.31, 6.73, 4.48, 3.91, 3.54, 2.8]
city = ['Сургут', 'Москва', 'Омск', 'Санкт-Петербург', 'Екатеринбург', 'Тюмень']
```

```python
width = 0.4
x = np.arange(len(city))
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, march, width, label='march')
rects2 = ax.bar(x + width/2, april, width, label='april')
ax.set_title('Доля пересечения по городам в марте-апреле')
ax.set_xticks(x)
ax.set_xticklabels(city, rotation=45, ha='right')
ax.legend()
```
